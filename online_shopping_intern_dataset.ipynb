# 2. Calculate the cumulative sales per month for each year.

```python
order_item  = pd.read_excel("/content/order_items.xlsx")     # 12 Question
order_item
```

```python
order  = pd.read_excel("/content/orders.xlsx")
order
```

```python
sales = order.merge(order_item[["order_id","freight_value","price"]],on = "order_id")
```

```python
sales["year"] = pd.to_datetime(sales["order_purchase_timestamp"]).dt.year
sales["month"] = pd.to_datetime(sales["order_purchase_timestamp"]).dt.month
```

```python
sales["total"] = sales["price"] + sales["freight_value"]
```

```python
monthly_sales  = sales.groupby(["year","month"])["total"].sum().reset_index()
monthly_sales
```

```python
monthly_sales["cumulative_sales"] = monthly_sales.groupby("year")["total"].cumsum()
```

```python
monthly_sales
```

```python
plt.figure(figsize=(10,10))
plt.plot(monthly_sales["cumulative_sales"])
plt.xlabel("Month")
plt.ylabel("Cumulative Sales")
plt.title("Cumulative Sales per Month")
plt.show()
```

#  3. Calculate the year-over-year growth rate of total sales.

```python
order_items = pd.read_excel("/content/order_items.xlsx")
order_items.head()
```

```python
orders = pd.read_excel("/content/orders.xlsx")
orders.head()
```

```python
orders["year"] = pd.to_datetime(orders["order_purchase_timestamp"]).dt.year   # 13 Question
```

```python
sales = orders.merge(order_items, on="order_id")
yearly_sales = sales.groupby("year")["price"].sum().reset_index()
yearly_sales["previous_year_sales"] = yearly_sales["price"].shift(1)
yearly_sales["yoy_growth_rate"] = round(
    ((yearly_sales["price"] - yearly_sales["previous_year_sales"])
     / yearly_sales["previous_year_sales"]) * 100, 2)
yearly_sales
```

```python
plt.plot(yearly_sales["year"], yearly_sales["price"])
plt.xticks(yearly_sales["year"])
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.title("Total Sales per Year")
plt.show()
```

#  4. Calculate the retention rate of customers, defined as the percentage of customers who make another purchase within 6 months of their first purchase.

```python
order.groupby("customer_id").filter(lambda x:len(x)>1)    # 14 Question
```

```python
first_purchase = orders.groupby("customer_id")["order_purchase_timestamp"].min().reset_index()
first_purchase.rename(columns = {"order_purchase_timestamp":"first_purchase_date"},inplace = True)
```

```python
merge = pd.merge(orders,first_purchase,on = "customer_id")
```

```python
merge["date_diff"] = (merge["order_purchase_timestamp"] - merge["first_purchase_date"]).dt.days
```

```python
repeat_customers = merge[(merge["date_diff"]>0) & (merge["date_diff"]<=180)]["customer_id"].nunique()
```

```python
total_customer = first_purchase["customer_id"].nunique()
```

```python
retained_rate = (repeat_customers/total_customer)*100
retained_rate
```

# 5. Identify the top 3 customers who spent the most money in each year.

```python
order = pd.read_excel("/content/orders.xlsx")   # 15 Question
order
```

```python
merge=  pd.merge(order,payment,on = "order_id")
merge["year"]  = pd.to_datetime(merge["order_purchase_timestamp"]).dt.year
customer_yearly_spending = merge.groupby(["customer_id","year"])["payment_value"].sum().reset_index()
customer_yearly_spending = customer_yearly_spending.sort_values(["year","payment_value"],
                                                                ascending = [True,False])
top_customers_per_year = customer_yearly_spending.groupby("year").head(3)
top_customers_per_year
customer_yearly_spending["rank"] = customer_yearly_spending.groupby("year")["payment_value"].rank(method="dense",
                                                                                                  ascending=False)
top_customers_per_year = customer_yearly_spending[
    customer_yearly_spending["rank"] <= 3]
top_customers_per_year
```
